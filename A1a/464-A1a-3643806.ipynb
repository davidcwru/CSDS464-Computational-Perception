**1a.** Use markdown and latex to write a short description of the univariate normal probability distribution function (pdf) $p(x | \mu, \sigma)$.


If a random variable $x$ follows a normal distribution

$$x \sim \mathcal{N}(\mu, \sigma)$$

where $\mu$ is the mean and $\sigma$ is the standard deviation, the likelihood of $x$ is given by the normal (or Gaussian) probability density function

$$p(x|\mu, \sigma) = \frac{1}{\sqrt{2\pi\sigma^2}} \exp\left(-\frac{1}{2} \frac{(x - \mu)^2}{\sigma^2}\right)$$


***


**1b.** 

For _N_ independent random variables

$$x_i \sim \mathcal{N}(\mu, \sigma), \quad i = 1 \ldots N$$

the joint probability distribution is given by

$$
\begin{aligned}
p(x_{1:N}|\mu, \sigma) &= \prod_{i=1}^{N} p(x_i|\mu, \sigma) \\
&= \frac{1}{(2\pi\sigma^2)^{\frac{N}{2}}} \exp\left(-\frac{1}{2} \sum_{i=1}^{N} \frac{(x_i - \mu)^2}{\sigma^2}\right)
\end{aligned}
$$

### 2. Simple functions and plotting


```julia
# General code used by both plots

using Statistics
using Plots

function g(x; μ = mu_value, σ = sigma_value)
    return 1 / (σ * sqrt(2 * π)) * exp(-0.5 * ((x - μ) / σ)^2)
end

function createPlot()
    
    y_vals  = g.(x_vals)
    p_at_x  = g(x_value)
    x_lower = x_value
    x_upper = x_value
    y_lower = 0
    y_upper = p_at_x

    plot(x_vals, y_vals, label="", lw=2)

    # Add a vertical line 
    plot!([x_lower, x_upper], [y_lower, y_upper], color=:red, lw=2, label="")

    # Add a dot
    scatter!([x_value], [p_at_x], label="p(x1 = $x_value | μ, σ) = $(round(p_at_x, digits=3))", color=:green, markersize=6)

    # Set plot title and display result
    title!("p(x | μ = $mu_value, σ = $sigma_value)")

end
```

```julia
# [1]

# Values for this Plot
mu_value    = 0;
sigma_value = 1;
x_value     = 1;
x_vals      = -4:0.01:4

# Create and display plot based on provided values
display(createPlot())

# Save this plot to a file
#savefig("464-A1a-3643806-plot1.png")

```
